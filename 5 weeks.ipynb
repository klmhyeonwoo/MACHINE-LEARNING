```python
import mglearn 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
```

```python
x, y = mglearn.datasets.make_wave(n_samples=50)

plt.plot(x, y, 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-3,3) # Y축 -3부터 3까지 증가
plt.show()
```

```python
train_input, test_input, train_target, test_target = train_test_split(x, y, test_size = 0.3, random_state = 42) # 결과 공유를 위해 랜덤 값을 고정된 상태로 제공합니다
xs = np.linspace(-3, 3, 1000) #  -3, 3까지 1000개의 요소를 1차원 배열로 생성합니다
```

```python
# K가 1일 경우
knr = KNeighborsRegressor(n_neighbors=1)
knr.fit(train_input, train_target)

plt.plot(xs, knr.predict(xs.reshape(-1,1)))
plt.plot(train_input, train_target, '^') # 훈련 데이터는 주황색 데이터
plt.plot(test_input, test_target, 'v') # 테스트 데이터는 초록색 데이터
plt.title(f'K : {knr.n_neighbors}, train score : {knr.score(train_input, train_target):.2f},  test score : {knr.score(test_input, test_target):.2f}')
plt.ylim(-3,3)
plt.show()
```

```python
# K가 3일 경우

knr = KNeighborsRegressor(n_neighbors=3)
knr.fit(train_input, train_target)

plt.plot(xs, knr.predict(xs.reshape(-1,1)))
plt.plot(train_input, train_target, '^') # 훈련 데이터는 주황색 데이터
plt.plot(test_input, test_target, 'v') # 테스트 데이터는 초록색 데이터
plt.title(f'K : {knr.n_neighbors}, train score : {knr.score(train_input, train_target):.2f},  test score : {knr.score(test_input, test_target):.2f}')
plt.ylim(-3,3)
plt.show()
```

```python
# K가 5일 경우, 테스트 당시 BEST K 인 것을 확인할 수 있다.
# 1의 경우 훈련세트와 테스트세트의 스코어 과대적합
# 3의 경우 훈련세트와 테스트세트의 스코어 과대적합
# 5의 경우 훈련세트와 테스트세트의 스코어 비교적 적합한 과대적합
# 7의 경우 훈련세트와 테스트세트의 스코어 과소적합
# 9의 경우 훈련세트와 테스트세트의 스코어가 같음
# 결과적으로, 1에 가까우면서 비교적 적합한 훈련세트와 테스트세트의 과대적합 점수를 가진 5의 경우(Hyper Parameter)가 최적의 값이라고 할 수 있다.

knr = KNeighborsRegressor()
knr.fit(train_input, train_target)

plt.plot(xs, knr.predict(xs.reshape(-1,1)))
plt.plot(train_input, train_target, '^') # 훈련 데이터는 주황색 데이터
plt.plot(test_input, test_target, 'v') # 테스트 데이터는 초록색 데이터
plt.title(f'K : {knr.n_neighbors}, train score : {knr.score(train_input, train_target):.2f},  test score : {knr.score(test_input, test_target):.2f}')
plt.ylim(-3,3)
plt.show()
```

```python
# K가 7일 경우
knr = KNeighborsRegressor(n_neighbors=7)
knr.fit(train_input, train_target)

plt.plot(xs, knr.predict(xs.reshape(-1,1)))
plt.plot(train_input, train_target, '^') # 훈련 데이터는 주황색 데이터
plt.plot(test_input, test_target, 'v') # 테스트 데이터는 초록색 데이터
plt.title(f'K : {knr.n_neighbors}, train score : {knr.score(train_input, train_target):.2f},  test score : {knr.score(test_input, test_target):.2f}')
plt.ylim(-3,3)
plt.show()
```
